US Cities Temperature Data 1995-2020 Extraction

```python
# Setup and Dependencies
import pandas as pd
import os
```

```python
# Creates DataFrame of the "Daily Temperature of Major Cities" data located on Kaggle
major_city_temps_file = "Resources/Daily Temperature of Major Cities/archive.zip"
major_city_temps_df = pd.read_csv(major_city_temps_file, compression='zip', header=0, sep=',', quotechar='"', dtype='unicode')
major_city_temps_df.head()
```

```python
# Isolates US Major City Data
us_temp_df = major_city_temps_df.loc[major_city_temps_df['Country'] == "US"]
us_temp_df.head()
```

```python
# Exports US Temperature Dataframe as a new CSV file called "US_Temp_1995_2020.csv"
file_path = 'Resources/'
us_temp_df.to_csv(os.path.join(file_path, 'US_Temp_1995_2020.csv'))
```
