```python
# Dependancies
import pandas as pd
import shutil
import os
```

```python
base_epa_file = "Resources/EPA_annual_conc_by_monitor/annual_conc_by_monitor_"

# Read the first csv file and put into our combined_epa_df
epa_file = base_epa_file + '1995.zip'
print('Loading file ' + epa_file)
print('-------------------------------------')
combined_epa_df = pd.read_csv(epa_file,compression='zip', header=0, sep=',', quotechar='"',dtype='unicode')


# Loop through the rest of the files and combine into combined_epa_df
for year in range(1996,2021):
    epa_file = base_epa_file + str(year) + '.zip'
    print('Loading file ' + epa_file)
    print('-------------------------------------')    
    epa_df = pd.read_csv(epa_file,compression='zip', header=0, sep=',', quotechar='"',dtype='unicode')
    combined_epa_df = pd.concat([combined_epa_df,epa_df])

# Output the DataFrame
combined_epa_df.head()
```

```python
extracted_epa_first_df = combined_epa_df[['Parameter Name', 'Year', 'Sample Duration', 'Pollutant Standard','Metric Used',
                                   'Method Name','Units of Measure','1st Max Value','1st Max DateTime','State Name','County Name','City Name']].copy()
extracted_epa_second_df = combined_epa_df[['Parameter Name', 'Year', 'Sample Duration', 'Pollutant Standard','Metric Used',
                                   'Method Name','Units of Measure','2nd Max Value','2nd Max DateTime','State Name','County Name','City Name']].copy()
extracted_epa_third_df = combined_epa_df[['Parameter Name', 'Year', 'Sample Duration', 'Pollutant Standard','Metric Used',
                                   'Method Name','Units of Measure','3rd Max Value','3rd Max DateTime','State Name','County Name','City Name']].copy()
extracted_epa_fourth_df = combined_epa_df[['Parameter Name', 'Year', 'Sample Duration', 'Pollutant Standard','Metric Used',
                                   'Method Name','Units of Measure','4th Max Value','4th Max DateTime','State Name','County Name','City Name']].copy()
```

```python
extracted_epa_first_df = extracted_epa_first_df.rename(columns = {'1st Max Value':'Max Value','1st Max DateTime':'Max DateTime'})
extracted_epa_second_df = extracted_epa_second_df.rename(columns = {'2nd Max Value':'Max Value','2nd Max DateTime':'Max DateTime'})
extracted_epa_third_df = extracted_epa_third_df.rename(columns = {'3rd Max Value':'Max Value','3rd Max DateTime':'Max DateTime'})
extracted_epa_fourth_df = extracted_epa_fourth_df.rename(columns = {'4th Max Value':'Max Value','4th Max DateTime':'Max DateTime'})
```

```python
combined_extracted_epa_df = pd.concat([extracted_epa_first_df,extracted_epa_second_df,extracted_epa_third_df,extracted_epa_fourth_df])
combined_extracted_epa_df.head()
```

```python
# Create the first zip file
compression_opts = dict(method='zip',
                        archive_name='US_EPA_1995_2020_first.csv')  
extracted_epa_first_df.to_csv('US_EPA_1995_2020_first.zip', index=False,
                              compression=compression_opts)  
```

```python
# Create the second zip file
compression_opts = dict(method='zip',
                        archive_name='US_EPA_1995_2020_second.csv')  
extracted_epa_second_df.to_csv('US_EPA_1995_2020_second.zip', index=False,
                              compression=compression_opts)  
```

```python
# Create the third zip file
compression_opts = dict(method='zip',
                        archive_name='US_EPA_1995_2020_third.csv')  
extracted_epa_third_df.to_csv('US_EPA_1995_2020_third.zip', index=False,
                              compression=compression_opts)  
```

```python
# Create the fourth zip file
compression_opts = dict(method='zip',
                        archive_name='US_EPA_1995_2020_fourth.csv')  
extracted_epa_fourth_df.to_csv('US_EPA_1995_2020_fourth.zip', index=False,
                              compression=compression_opts)  
```

```python
# Move the zip files to Resources
curr_dir = os.getcwd()
files = ['US_EPA_1995_2020_first.zip','US_EPA_1995_2020_second.zip','US_EPA_1995_2020_third.zip','US_EPA_1995_2020_fourth.zip']

for zip_file in files:

    source = curr_dir + '/' + zip_file
    destination = curr_dir + '/Resources/' + zip_file
    # Move the content of 
    # source to destination 
    dest = shutil.move(source, destination) 
```
